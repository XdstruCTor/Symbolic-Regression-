***The SR algorithm includes cos because:***

**Search space definition***: We explicitly defined cos in the unary_operators, so the algorithm considers it during the search.

***Bias toward explainability***: While cos adds interpretability (a known mathematical function), its presence in the equation doesn't guarantee it's the best model—it depends on the problem and data.

```julia
#Pkg.add("SymbolicRegressio")
Pkg.add("MLJ")
import SymbolicRegression: SRRegressor
import MLJ: machine, fit!, predict, report

# Dataset with two named features:

X = (a= rand(500), b = rand(500))

# and one target:
y = @. 2 * cos(X.a * 23.5) - X.b^2

# with some noise:

y = y .+ randn(500) .* 1e-3
model = SRRegressor(
    niterations = 50,
    binary_operators=[+,-,*],
    unary_operators=[cos],
)
```

```julia
mach = machine(model, X, y)
fit!(mach)
```

## Interpretig the table


**Complexity**: The number of nodes (or terms) in the expression. Simpler expressions have lower complexity.

**Loss**: The mean squared error (MSE) between the predicted output (y) and the target values. Lower loss indicates a better fit.

**Score**: A metric that balances loss and complexity. Higher scores favor simpler yet accurate solutions.

**Equation**: The candidate expression generated by the SR algorithm

```julia
report(mach)
```

```julia
predict(mach, X)
```

## low-level Interface

The heart of SymbolicRegression.jl is the equation_search function. This takes a 2D array and attempts to model a 1D array using analytic functional forms. Note: unlike the MLJ interface, this assumes column-major input of shape

```julia
import SymbolicRegression: Options, equation_search

X= randn(2, 100)
y = 2* cos.(X[2, :]) + X[1, :] .^ 2 .-2

options = Options(
    binary_operators=[+,*,/,-],
    unary_operators =[exp],
    populations = 20
)

hall_of_fame = equation_search(
    X, y, niterations=40, options=options,
    parallelism=:multithreading
)
```

### The exp does Not Appear Frequently because of

**Data-Driven Selection**:

Symbolic Regression algorithms prioritize operators that minimize the loss on the dataset.
If `exp` doesn't significantly reduce the loss compared to simpler operators like multiplication, addition, or subtraction, it will be less frequently used.

**Search space Exploration**: Even though `exp` is included, the algorithm explores a vast space of possible expressions. If the dataset's patterns don't align well with exponential relationships, exp will have a lower chance of being included.

**Complexity**: Measured by the number of operations, variables, or terms in the equation.
**Loss**: Quantifies how well the model fits the data (e.g., mean squared error).


```julia
import SymbolicRegression: calculate_pareto_frontier

dominating = calculate_pareto_frontier(hall_of_fame)
```

Dominating Models:
A model is dominant if the model achieves both:

Lower loss.
Lower or equal complexity.
For instance:

Complexity `3` is dominate complexity 1.
Complexity `3` is dominated by complexity `5`

**Start Simple**:

`Complexity 1`: 𝑦=−0.064224 is the simplest possible solution but has the highest loss (1.858). It’s fast to compute but provides poor accuracy.

**Trade-offs Emerge**:

`At complexity 3`: 
y=x1 ∗x1​ introduces non-linearity, reducing loss to 1.560.
`At complexity 5`: 

y=(x1 ∗x1 )+−0.83855 adds a constant term, further improving loss to 0.857.

**High Complexity, Low Loss**:

At `complexity 16`: The equation incorporates exp, leading to significantly lower loss (0.02936). However, it’s more complex

```julia
trees = [member.tree for member in dominating]
```

```julia
tree = trees[end]
output = tree(X)
```

## constructing expressions

The Expression type wraps this and includes metadata about operators and variable names.

You can manipulate and construct expressions directly.

Defining a specific equation (e.g., cos(x1 - 3.2 * x2) - x1^2) serves two purposes:

**Validation**: You might want to test a hypothesized relationship to see how well it performs on the data.
**Initialization**: In symbolic regression, such trees can serve as starting points or baselines for optimization.
By comparing your custom equation (tree) with the results on the Pareto frontier, you can:

See if the algorithm discovers better equations.
Validate whether your equation is optimal or dominated by others.

If we remove tree and rely solely on symbolic regression, the algorithm will start with random equations and iteratively optimize them using evolutionary algorithms.

```julia
using SymbolicRegression: Options, Expression, Node

options = Options(;
    binary_operators=[+, -, *, /], unary_operators=[cos, exp, sin]
)

operators = options.operators
variable_names = ["x1", "x2", "x3"]

x1, x2, x3 = [Expression(Node(Float32; feature=i); operators, variable_names) for i=1:3]
tree = cos(x1 - 3.2 * x2) - x1 * x1
```

`did_succeed` is a Boolean indicator (or part of a result tuple) that tells you whether the evaluation of a tree expression for a given input dataset X was successful.

For example: If the tree includes a division operator `(/)`, there’s a risk that:

The denominator evaluates to zero at some point during the computation for a given set of inputs.
For example, if the tree has a term like $x_1$ / ($x_2$ - 2.5) and $x_2$ = 2.5, this would lead to division by zero.
Dividing by zero is mathematically undefined, so the algorithm must handle this case gracefully to avoid crashing.

```julia
#float32_tree = convert(Expression{Float32}, tree)
using SymbolicRegression

X= rand(Float32, 3, 100)
tree(X)
output, did_succeed = SymbolicRegression.eval_tree_array(tree, X)
dominating = calculate_pareto_frontier(hall_of_fame)
```

```julia
import SymbolicRegression: compute_complexity, string_tree

println("Complexity\tMSE\tEquation")

for member in dominating
    complexity = compute_complexity(member, options)
    loss = member.loss
    string = string_tree(member.tree, options)
    
    println("$(complexity)\t$(loss)\t$(string)")
end
```

```julia
complexities = []
losses = []

for member in dominating
    push!(complexities, compute_complexity(member, options))
    push!(losses, member.loss)
end
```

```julia
# Scatter plot of complexity vs. loss
scatter(complexities, losses,
    xlabel="Complexity",
    ylabel="Loss",
    title="Pareto Frontier",
    marker=:circle,
    legend=false)
```
